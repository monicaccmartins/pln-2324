```python
f = open('medicina.json', 'r', encoding='UTF-8')
medicina = json.load(f)

ii = 0

dic_es = {}
dic_en = {}
dic_pt = {}

for entrada in medicina:   # para cada indice
    dic_es[entrada] = {}
    dic_en[entrada] = {}
    dic_pt[entrada] = {}
    dic_es[entrada]['Traducoes'] = {'en': medicina[entrada]["Traduções"]['en'], 'pt': medicina[entrada]["Traduções"]['pt']}
    dic_en[entrada]['Traducoes'] = {'es' : medicina[entrada]["Traduções"]['es'], 'pt': medicina[entrada]["Traduções"]['pt']}
    dic_pt[entrada]['Traducoes'] = {'en': medicina[entrada]["Traduções"]['en'], 'es' : medicina[entrada]["Traduções"]['es']}
    dic_es[entrada]["Termo"] = medicina[entrada]["Traduções"]["es"]
    dic_en[entrada]["Termo"] = medicina[entrada]["Traduções"]["en"]
    dic_pt[entrada]["Termo"] = medicina[entrada]["Traduções"]["pt"]
    dic_es[entrada]["Fontes"] = ['Vocabulário de Medicina']
    dic_en[entrada]["Fontes"] = ['Vocabulário de Medicina']
    dic_pt[entrada]["Fontes"] = ['Vocabulário de Medicina']

    if "Categoria gramatical" in medicina[entrada]: # se houver ent há area de aplicação, pois pertence ao medicina
        dic_es[entrada]["Categoria gramatical"] = medicina[entrada]["Categoria gramatical"] # não faz sentido equiparar

        areas_es = medicina[entrada]["Área(s) de aplicação"]
        dic_es[entrada]["Área(s) de aplicação"] = areas_es
        dic_en[entrada]["Área(s) de aplicação"] = [str(GoogleTranslator(source="es", target="en").translate(area)) + ' * ' for area in areas_es]
        dic_pt[entrada]["Área(s) de aplicação"] = [str(GoogleTranslator(source="es", target="pt").translate(area)) + ' * ' for area in areas_es]

    if "SIN" in medicina[entrada]:
        Sin_es = medicina[entrada]["SIN"]
        dic_es[entrada]["Sinonimos"] = [Sin_es]
        dic_en[entrada]["Sinonimos"] = [str(GoogleTranslator(source="es", target="en").translate(Sin_es)) + ' * ']
        dic_pt[entrada]["Sinonimos"] = [str(GoogleTranslator(source="es", target="pt").translate(Sin_es)) + ' * ' ]
        
    if "VAR" in medicina[entrada]:
        dic_es[entrada]["Variante"] = medicina[entrada]["VAR"] # não faz sentido traduzir
    
    if 'Entrada Remissiva' in medicina[entrada]:
        dic_es[entrada]["Entradas"] = {}
        dic_en[entrada]["Entradas"] = {}
        dic_pt[entrada]["Entradas"] = {}
        for rem in medicina[entrada]['Entrada Remissiva']:
            vid = medicina[entrada]['Entrada Remissiva'][rem].replace("Vid.- ","")
            remen = GoogleTranslator(source="es", target="en").translate(rem)
            rempt = GoogleTranslator(source="es", target="pt").translate(rem)
            viden = GoogleTranslator(source="es", target="en").translate(vid)
            vidpt = GoogleTranslator(source="es", target="pt").translate(vid)
            dic_es[entrada]["Entradas"][rem] = vid
            dic_en[entrada]["Entradas"][remen] = viden
            dic_pt[entrada]["Entradas"][rem] = vidpt

    if (len(list(dic_en.keys())) % 100) == 0:                   # a cada 100 gravar

        f_out_es = open("doc_conc_es.json","w", encoding='UTF-8')
        f_out_en = open("doc_conc_en.json","w", encoding='UTF-8')
        f_out_pt = open("doc_conc_pt.json","w", encoding='UTF-8')

        json.dump(dic_es,f_out_es,indent=4, ensure_ascii=False)
        json.dump(dic_en,f_out_en,indent=4, ensure_ascii=False)
        json.dump(dic_pt,f_out_pt,indent=4, ensure_ascii=False)

        f_out_es.close()
        f_out_en.close()
        f_out_pt.close()


    print(ii, entrada)
    ii+=1

f_out_es = open("doc_conc_es.json","w", encoding='UTF-8')
f_out_en = open("doc_conc_en.json","w", encoding='UTF-8')
f_out_pt = open("doc_conc_pt.json","w", encoding='UTF-8')

json.dump(dic_es,f_out_es,indent=4, ensure_ascii=False)
json.dump(dic_en,f_out_en,indent=4, ensure_ascii=False)
json.dump(dic_pt,f_out_pt,indent=4, ensure_ascii=False)


f_out_es.close()
f_out_en.close()
f_out_pt.close()
```

```python
f_es = open("doc_conc_es_V_GMS_outros.json","r", encoding='UTF-8')
f_en = open("doc_conc_en_V_GMS_outros.json","r", encoding='UTF-8')
f_pt = open("doc_conc_pt_V_GMS_outros.json","r", encoding='UTF-8')

es = json.load(f_es)
en = json.load(f_en)
pt = json.load(f_pt)

f_es.close()
f_en.close()
f_pt.close()


def concatena(d, repetidos):
    if not repetidos:
        return d
    indice_base = list(repetidos.keys())[0]
    conteudo_base = d[indice_base]
    for indice, conteudo in repetidos.items():
        if indice != indice_base:
            for chave, info in conteudo.items():
                if chave in conteudo_base:
                    if isinstance(conteudo_base[chave], str) and chave != 'Termo':
                        conteudo_base[chave] += ' ' + info
                    elif isinstance(conteudo_base[chave], list):
                        temp_set = set(conteudo_base[chave])
                        temp_set.update(info)
                        conteudo_base[chave] = list(temp_set)
                    elif isinstance(conteudo_base[chave], dict): # Por exemplo Traduçoes : {'en' : '...', 'pt' : '...'}
                        for c, v in info.items():
                            if c in conteudo_base[chave]:
                                if isinstance(conteudo_base[chave][c], str):
                                    if v not in conteudo_base[chave][c]:
                                        conteudo_base[chave][c] += ', ' + v
                            else:
                                conteudo_base[chave][c] = v
                else:
                    conteudo_base[chave] = info
            del d[indice]
    d[indice_base] = conteudo_base


def getTermosIguais(dicionario, termo):
    repetidos = {}
    for indice in dicionario:
        if dicionario[indice]['Termo']:
            if dicionario[indice]['Termo'].lower() == termo:
                repetidos[indice] = dicionario[indice]
    return repetidos

def findIguais(dicionario):
    termosIndice = [(dicionario[indice]['Termo'].lower() , indice) for indice in dicionario if dicionario[indice]['Termo']]
    termos = [dicionario[indice]['Termo'].lower() for indice in dicionario if dicionario[indice]['Termo']]
    for termo in termosIndice:
        t, i = termo
        numeroVezes = termos.count(t)
        if numeroVezes > 1:
            repetidos = getTermosIguais(dicionario, t)
            concatena(dicionario, repetidos)
        
    return dicionario

en = findIguais(en)
es = findIguais(es)
pt = findIguais(pt)

f_out_es = open("doc_conc_es_V_GMS_outros_semRepetidos.json","w", encoding='UTF-8')
f_out_en = open("doc_conc_en_V_GMS_outros_semRepetidos.json","w", encoding='UTF-8')
f_out_pt = open("doc_conc_pt_V_GMS_outros_semRepetidos.json","w", encoding='UTF-8')

json.dump(es,f_out_es,indent=4, ensure_ascii=False)
json.dump(en,f_out_en,indent=4, ensure_ascii=False)
json.dump(pt,f_out_pt,indent=4, ensure_ascii=False)

f_out_es.close()
f_out_en.close()
f_out_pt.close()
```
